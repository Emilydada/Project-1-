```python
import pandas as pd
import os
```

```python
file = "voter_summary2015.csv"
```

```python
original_df = pd.read_csv(file)
original_df.head(300)
```

```python
count = original_df["Age group"].value_counts()
count.head()
```

```python
#Remove the columns of some data columns that no need for calculations in age and sex groups.
cleaned_list = original_df.drop(columns=[ 'UOM','COORDINATE','SCALAR_ID','VECTOR','STATUS','SYMBOL','TERMINATED','SCALAR_FACTOR','Statistics','DECIMALS'])
cleaned_list.head()
```

```python
grouped_voter_data = cleaned_list.groupby(['Sex', 'Age group'])
grouped_voter_data.head(400)
```

```python
#bar chart of the total female and male's voters 
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
sex_summary=grouped_voter_data['VALUE'].sum().head(200)
sex_summary.head(200)
```

```python
w=0.4
x_axis=["25 to 34 years"," 35 to 44 years ","45 to 54 years ","  55 to 64 years","65 to 74 years","75 years and over"]
voter_count1=[6260,6150.3,6872.2,5847.8,4099,3536.2]
voter_count2=[6153.2,5934.8,6856.1,5604.8,3812.8,2649.6]

x_axis1 = np.arange(len(voter_count1))
x_axis2=[i+w for i in x_axis1]
```

```python
plt.bar(x_axis1, voter_count1 , w,color="lightcoral", label="Female")
plt.bar(x_axis2, voter_count2, w,color="lightblue", label="Male" )
```

```python
tick_locations = [value for value in x_axis1]
plt.xticks(tick_locations, x_axis, rotation=12)
```

```python
plt.title("Voter Sex and Age Group Distribution 2015")
plt.xlabel("Voter Sex and Group")
plt.ylabel("Voter Total")
```
