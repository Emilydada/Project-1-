```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
w=0.4
status=["Canadian-born,non-Aboriginal persons", "Aboriginal people", "Immigrants"]
turnout_rate=[70.9,53.5,68.5]
turnout_rate2=[77.9,68.1,74.9]
x_axis1 = np.arange(len(turnout_rate))
x_axis2=[i+w for i in x_axis1]
```

```python
plt.bar(x_axis1, turnout_rate , w,color="lightcoral", label="2011")
plt.bar(x_axis2, turnout_rate2, w,color="lightblue", label="2015" )
```

```python
tick_locations = [value for value in x_axis1]
plt.xticks(tick_locations, status, rotation=12)
```

```python
plt.title("Voter Turnout Rate 2011 vs. 2015")
plt.xlabel("Voter Status")
plt.ylabel("Voter Turnout Rate")
```
